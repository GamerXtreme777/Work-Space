```python
# Making a sctter plot to describe the data
plt.figure(figsize=(8, 6))
plt.scatter(dataset['fixed acidity'], dataset['pH'], alpha=0.6, color='b')
plt.title('Scatter Plot of Fixed Acidity vs. pH')
plt.xlabel('fixed acidity')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```

```python
# Making histogram to describe the dataset
plt.hist(dataset['fixed acidity'], bins=10)  
plt.xlabel('fixed acidity')
plt.ylabel('frequency')
plt.title('Histogram')
plt.show()
```
